```python
###   Операции сравнения
#     ==  равно
#     !=  не равно
#     <   меньше
#     >   больше
#     <=  меньше или равно
#     >=  больше или равно 
```

```python
# В Питоне допустимы и логические выражения, содержащие несколько знаков сравнения, например x < y < z. 
# При этом все сравнения обладают одинаковым приоритетом, который меньше, чем у любой арифметической операции.

# Результат вычисления логического выражения можно сохранять в переменную, которая будет иметь тип bool. 
# Переменные такого типа, как и числа и строки, являются неизменяемыми объектами.
```

```python
# Логические операции

# Чтобы записать сложное логическое выражение, часто бывает необходимо воспользоваться логическими связками 
# "и", "или" и "не" - and, or и not 
# соответственно. 
# Операции and и or являеются бинарными, т.е. должны быть записаны между операндами, например x < 3 or y > 2. 
# Операция not - унарная и должна быть записана перед единственным своим операндом.

# Все логические операции имеют приоритет ниже, чем операции сравнения (а значит, и ниже чем арифметические операции).

# Приоритет логических операций: 
#     1 - not,(наивысший приоритет)  
#     2 - and 
#     3 - or (наименьший приоритет) 
# На порядок выполнения операций можно влиять с помощью скобок, как и в арифметических выражениях.
```

```python
# Пример1:

# Проверка на четность (остаток отделения на 2 равен 0)
n = int(input())
isEven = n % 2 == 0
print(isEven)
```

```python
# Привмер2:
# Пересекаются ли между собой два отрезка времени или просто два отрезка, находящихся вдоль одной прямой?
# Два отрезка пересекаются если или начало или конец одного из них находится внутри дугого!
# Введем 4 перкменные - начало и конец для каждого отрезка:

start1 = int(input()) #для простоты вводим простые числа (1,2... 3, 20)
finish1 = int(input())
start2 = int(input())
finish2 = int(input())
# Проверяем все варианты
isS1in2 = start2 <= start1 <= finish2 # начало первого отрезка находится внутри второго отрезка
isF1in2 = start2 <= finish1 <= finish2 # конец первого отрезка находится внутри второго отрезка
isS2in1 = start1 <= start2 <= finish1 # начало второго отрезка находится внутри первого отрезка
isF2in1 = start1 <= finish2 <= finish1 # конец первого отрезка находится внутри первого отрезка
answer = isS1in2 or isF1in2 or isS2in1 or isF2in1 # если хотя бы одно условие истинно - отрезки пересекаются
print(answer)
# Все эти условия можно записать(описать) короче
# начало первого отрезка меньше или равно, чем конец второго и доновременно 
# начало второго меньше или равно, чем конец первого
answer = start1 <= finish2 and start2 <= finish1
print(answer)
```
